```python
from sklearn.model_selection import train_test_split
```

```python
#converting string date to date-time
import datetime as dt
pd.options.display.float_format = '{:.2f}'.format
india_case['date']=pd.to_datetime(india_case['date'])
india_case
```

```python
#converting date to ordinal values
india_case['date']=india_case['date'].map(dt.datetime.toordinal)
india_case
```

```python
X=india_case.drop(['iso_code','continent','location','total_cases'],axis=1)
Y=india_case['total_cases']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1234)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
print('R-square score',r2_score(Y_test,y_pred))
```

```python
adj_rsquares= 1 - (1 - r2_score(Y_test, y_pred)) * ((len(Y_test) - 1)/(len(Y_test)-X_train.shape[1]-1))
print('Adjusted R-squared score',adj_rsquares)
```
